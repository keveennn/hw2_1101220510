pilih hanya kolom numerik dan Hitung Q1, Q3, dan IQR

```python
kv_numeric = kv.select_dtypes(include=['number'])
Q1 = kv_numeric.quantile(0.75)
Q3 = kv_numeric.quantile(0.25)
IQR = Q3 - Q1
```

buang outlier pada kolom numerik

```python
kv_numeric_cleaned = kv_numeric[~((kv_numeric < (Q1 - 1.5 * IQR)) | 
(kv_numeric > (Q3 + 1.5 * IQR))).any(axis=1)]
kv_cleaned = kv.loc[kv_numeric_cleaned.index]
print(kv_cleaned.shape)
```

menampilkan jumlah mobil sport berdasarkan harga

```python
kv.Model.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Harga Mobil berdasarkan Model")
plt.ylabel('Harga Mobil')
plt.xlabel('Model');
```
